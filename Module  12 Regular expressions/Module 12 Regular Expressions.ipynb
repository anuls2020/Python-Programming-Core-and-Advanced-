### <font color='#3498DB'>Programming Examples</font>

<font color='#D2691E'>**Qn) Write a Python Program for searching the keyword "Python from the given data"**<br>


```python
# RegEx1.py
# This program searches the word "Python" in the given data. also it's occurrence
import re

givendata = "Python is an OOP language and Python is also a functional programming language"

reg = "Python"

mattab = re.finditer(reg, givendata)
print("Type of mattab:", type(mattab))
print(mattab)
print("-" * 60)
noc = 0
for val in mattab:
    print("start index = {}, End index = {}, Matched Value = {}".format(val.start(), val.end(), val.group()))
    noc += 1
print("-" * 60)
print("Number of occurrence of word {} is {}".format(reg, noc))
```

<font color='#D2691E'>**Qn) Write a Python Program for searching the keyword "Python from the given data and find number of occurrence"**<br>


```python
# RegEx2.py
# This program searches the word "Python" in the given data and also find it's occurrence
import re

givendata = "Python is an OOP language and Python is also a functional programming language"

reg = "Python"

mattab = re.finditer(reg, givendata)
print("Type of mattab:", type(mattab))
print(mattab)
print("-" * 60)
noc = 0
for val in mattab:
    print("start index = {}, End index = {}, Matched Value = {}".format(val.start(), val.end(), val.group()))
    noc += 1
print("-" * 60)
print("Number of occurrence of word {} is {}".format(reg, noc))
```

<font color='#D2691E'>**Qn) Write a Python Program for searching the keyword "Python from the given data and find number of occurrence"**<br>


```python
# RegEx3.py
#

import re

data = "Python is a good programming language. Python supports object oriented programming. Python supports regular expressions"

reg = "Python"

matchdata = re.findall(reg, data)

print("Type of matchdata", type(matchdata))
print(matchdata)
print("Number of occurrence of the word {} is {}".format(reg, len(matchdata)))
```

<font color='#D2691E'>**Qn) Write a Python Program for searching the keyword "Python from the given data and find number of occurrence"**<br>


```python
# RegEx4.py
# This program searches for a word "Python" in given data and give its first occurrence

import re

data = "Python is a good programming language. Python supports object oriented programming. Python supports regular expressions"

reg = "Python"

matchdata = re.search(reg, data)

print("Type of First occurrence of the regular expression", type(matchdata))
if (matchdata != None):
    print(matchdata)
else:
    print("The word {} is not found on the given data".format(reg))
```

### <font color='#3498DB'>Programmer-Defined Character Classes</font>

* Programmer-Defined Character Classes are prepared by Programmers and they **helps us to prepare search patterns and they are used to search in the given data for obtaining desired result**.


<div class="alert alert-block alert-success">
<b>Syntax:</b> 
    
    [Search Pattern-Programmer-Defined Character Classes ]


* We can create 14  Programmer-Defined Character Classes. They are

    1. **[abc]**    --->It searches for **either 'a' or 'b' or 'c' only**
    2. **[^abc]**   --->It searches for all **except  'a' or 'b' or 'c'**
    3. **[a-z]**    --->It searches for all **lower case alphabets only**
    4. **[^a-z]**   --->It searches for all **except lower case alphabets**
    5. **[A-Z]**    --->It searches for all **Upper case alphabets only**
    6. **[^A-Z]**   --->It searches for all **except Upper case alphabets**
    7. **[0-9]**    --->It searches for all **digits only**
    8. **[^0-9]**   --->It searches for all **except digits**
    9. **[a-z 0-9]**   --->It searches for **lower case alphabets and digits only**
    10. **[^a-z 0-9]** --->It searches for all **except lower case alphabets and digits**
    11. **[A-Z a-z]**  --->It searches for all **alphabets (lower and upper) only**
    12. **[^A-Z a-z]** --->It searches for all **except alphabets (lower and upper) only**
    13. **[A-Z a-z 0-9]**  --->It searches for all **alphabets (lower and upper) and digits (except special symbols)**
    14. **[^A-Z a-z 0-9]** --->It searches for all **special symbols**
</div>    

### <font color='#3498DB'>Pre-Defined Character Classes</font>

* **Pre-Defined Character Classes** are already defined python software and they helps us to prepare search patterns and they are used to search in the given data for obtaining desired result.

<div class="alert alert-block alert-success">
<b>Syntax:</b> 
    
`\Pre-defined character class`


* We have  6 pre-defined Pre-Defined Character Classes. They are

	1. **\s** ----->It searches for **space character only**                           
	2. **\S**----->It searches for **all except space character**
	3. **\d**----->It searches for **digit only (or)  [ 0-9]**
	4. **\D**---->It searches for **all except digit (or) [^0-9]**
	5. **\w**---->It searchers for **word character (or ) [A-Za-z0-9]**
	6. **\W**---->It searcher for **special symbols (or) [^A-Za-z0-9]**

</div>
    

### <font color='#3498DB'>Quantifiers in Regular Expressions</font>

* Quantifiers **specify how many instances of a character, group, or character class must be present in the input for a match to be found**

(**OR**)

* Quantifiers in Regular Expression makes us to **understand the number of occurrence of search pattern searching in given data for obtaining desired result**.

* We have 5 types of quantifiers, they are
<div class="alert alert-block alert-success">
<b>Syntax:</b>   
    
 *  "**searchobj**"  :  Example: "a" --> searches **only one 'a'**
 *  "**searchobj+**" :  Example: "a+"--> searches **one 'a' or more 'a''s**
 *  "**searchobj***" :  Example: "a*"--> searches **zero 'a' or one 'a' or more 'a''s**
 *  "**searchobj?**  :  Example: "a?"--> searches for **zero 'a' or one 'a'**
 *  "**.**"          :  Example: "." --> searches for **all values in given data**
</div> 
 
<div class="alert alert-block alert-info">
<b>Tip/Note:</b> 
    
* **\d+**     -- [0-9]+      --> Searches for one more digits
* **\w+**     -- [a-zA-Z]+   --> Searches for one alphabet or more
* **\ddddd**  -- d{5}        --> Searches for 5 digit number
* **\dd.\dd** -- \d{2}.\d{2} --> Searches for 2 indigers and 2 decimals
</div>
