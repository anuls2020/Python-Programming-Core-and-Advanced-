## Inner (or) Nested list

- The process of defining one list inside another list is called inner or nested list.  

##### Syntax:-
<code><font color='#D2691E'>**listobj=[val1,val2....[ val11,val12,...] , [val21,val22....] ....valn-1]**</font></code>

- **[val11,val12,...]** is called one inner list
- **[val21,val22....]** is called another inner list
- **[val1,val2.... ....valn-1 ]** is called outer list
- On the list and inner list we can apply all operation regarding **Indexing, Slicing and all pre-defined functions**.

**Requirement:**

My Requirement -->To store

		stno----------10
		name -----Rossum
		Internal Marks of three subs---16,19,15
		External Marks of three subs---78,65,79
		College name---NLU

**Solution:**

<code>stlist=[10,"Mahesh",16,19,15,78 65,79,"NLU"]</code>
			
            (OR)
            
<code>stlst1=[ 10,"Mahesh", [16,19,15],[78,65,79], "NLU" ]</code>
