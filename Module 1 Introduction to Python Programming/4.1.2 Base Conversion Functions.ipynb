# Base Conversion Functions

* The purpose of Base Conversion Functions is that to **Convert One Base value into another base value**.
* In Python , we have **3 Base Conversion Functions**. They are
	* **bin()**
	* **oct()**
	* **hex()**

### 1. bin()

* This Function is used for **converting any type of base value into binary number system value**.
       
   <div class="alert alert-block alert-success">
   <b>Syntax:</b> <br>
    <code>varname=bin(decimal/octal/hexa decimal value)</code>
   </div>   

##### Examples:

```python
a=15
print(a,type(a))
b=bin(a)
print(b,type(b))
```

```python
a=0o14
print(a,type(a))
b=bin(a)
print(b,type(b))
```

```python
a=0xA
print(a,type(a))
b=bin(a)
print(b,type(b))
```

### 2. oct()

* This Function is used for **converting any type of base value into octal number system value.**

   <div class="alert alert-block alert-success">
   <b>Syntax:</b> <br>
        <code>varname=oct(decimal/binary/hexa decimal value)</code>
   </div>   

##### Examples:

```python
a=12
print(a,type(a))
b=oct(12)
print(b,type(b))
```

```python
a=0b1111
print(a,type(a))
b=oct(a)
print(b,type(b))
```

```python
a=0XACC
print(a,type(a))
b=oct(a)
print(b,type(b))
```

### 3. hex() 

* This Function is used for **converting any type of base value into hexa Decimal number system value**.

   <div class="alert alert-block alert-success">
   <b>Syntax:</b> <br>
        <code>varname=hex(decimal/binary/octal value)</code>
   </div>   

##### Examples:

```python
a=2764
print(a,type(a))
b=hex(a)
print(b,type(b))
```

```python
b=hex(15)
print(b,type(b))
```

```python
a=0o15
print(a,type(a))
b=hex(a)
print(b,type(b))
```

```python
a=0b1010
print(a,type(a))
b=hex(a)
print(b,type(b))
```
