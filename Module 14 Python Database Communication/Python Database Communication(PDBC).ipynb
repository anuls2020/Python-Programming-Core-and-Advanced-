## <font color='#3498DB'>**Limitations of Files**</font>

* We achieved the data persistency by using files. But the concept of files having the follwoing limitations.
  
  1. **File concept** of any language are **un-secured** because files concept unable to provide security in the form of **User Name and password**.
  2. The data of Files **does not contain column names**, so that it is **very difficult to process the data**. In otherwords, **Processing and Selecting the data** from files is very difficult.
  3. The files concept is **unable to store large volume of data**.
  4. The **architecture of files may differ from One OS to another OS**.

* To over come these limitations of files , we use another concept called <font color='#009900'>**DataBase softwares**</font>.

## <font color='#3498DB'>**Steps for Developing PDBC application with Oracle**</font>

* To develop any Python database communication Applications, we must use the following steps. 
  1. `import  cx_Oracle`
  2. Every Python must establish the connection from Oracle databse
  3. Create an object of `Cursor`.
  4. Design the query(It is a request / Question to the data base) and place it in an object of Cursor and execute.
  5. Python Program Process the result of the Query. 
  6. Python Program closes the connection.
  
**Detailed Explenation:**  

1. <font color='#009900'>**import  cx_Oracle**</font>
    * If a PYTHON program want to communicate with Oracle Database then we must import a pre-defined module called "cx_Oracle".
    * Example: `import  cx_Oracle`
2. <font color='#009900'>**Every Python script must establish the connection from Oracle database**</font> 
    * If a Python Programmer want to perform different database operations then first we must get a connection from Oracle database. 
    * To get the connection from Oracle Database, we must use a pre-defined function called "**connect()**" , which is present in **cx_Oracle** module and it returns an object of **<class, 'cx_Oracle.Connection'>**

        <div class="alert alert-block alert-success">
        <b>Syntax:</b><br> 
        <code><b>varname= cx_Oracle.connect("username/password@DNS/serviceid")</b></code>
        </div> 

    * Here "**username/password@DNS/serviceid**" is called **Connection URL**
    	* here "**username**" repersents **user name of Database**
        * here **password** represents the **password of the  Database**
        * Here **"DNS"(Domain Naming Service)** represents **Name of the machine** where the **Database software installed**.
        * The default DNS of every computer is "**localhost**"
        * Here In Place of DNS, **we can also use IP Address**( Internet Protocal Address) and it always Address of the machine where Database software installed.
        * The default IP address of every computer is 127.0.0.1(loop back address)
        * "**serviceid**" represents **on which name Oracle database is available in current working machine**.
        * here "**varname**" represents an **object of <class, 'cx_Oracle.Connection'>**
        
        <div class="alert alert-block alert-info">
        <b>Note:</b><br>
        To find serviceid of Oracle database, we must use the following. 
        <br>
        <code>SQL> select * from global_name;
        GLOBAL_NAME
        `-----------`
        ORCL
        `-----------`</code></div> 
  
  * <b>Examples:</b>  
        * **con = cx_Oracle.connect("anu/tiger@localhost/orcl")**
						(OR)
        * **con = cx_Oracle.connect("anu/tiger@127.0.0.1/orcl")**
        * print("Python Program obtained Connection from Oracle DB")
        * **Note**:- If we enter wrong Connection URL then we get an exception called "**cx_Oracle.DatabaseError**".
        

3. <font color='#009900'>**Create an object of Cursor:**</font>
    * "**Cursor**" is a pre-defined class present in cx_Oracle Module.
    * The purpose of Cursor class object is that "**It Carry the Query from Python Program and brings the result from Oracle Database and Hand Over to Python Program**. Hence an object of Cursor **acts as a driver / mediator / middleware layer between Python program and Database Software**.
    * Programatically, To create an object of Cursor class, we use **cursor()**, which is present in a **class <class, 'cx_Oracle.Connection'>**.
        
        <div class="alert alert-block alert-success">
        <b>Syntax:</b><br> 
        <code><b>varname=connobject.cursor()</b></code>
        </div> 
    
    * here "varname" is an object of class <class,"cx_Oracle.Cursor">

      * anucur=anucon.cursor()
	  * print("Cursor object Created")
	  * print("type of anucur=",type(anucur))---<class,"cx_Oracle.Cursor">


4. <font color='#009900'>**Design the query(It is a request / Question to the data base) and place it  in an object of Cursor and execute**</font>  
    * A Query is a **request / Question to the database from Python Program** to perform Certain database operation.
    * After designing the Query and  we must execute by using **execute()**, which is present in <class,"cx_Oracle.Cursor">
        <div class="alert alert-block alert-success">
        <b>Syntax:</b><br>

        </div>

### Types of Queries in Database Softwares

* A Query is a request / Question to the database from Python Program to perform Certain database operation.
* In Database Softwares, we 3 types of Queries. They are
  
  1. **DDL (Data Definition Language) Queries** <font color='#009900'>**(Never participates in rollback or commit)**</font>
    * create
    * alter ( add , modify)
    * drop
  2. **DML (Data Manipulation Language) Queries** <font color='#009900'>**(rollback and commit applicable only for DML statements)**</font>
    * insert
    * delete
    * update
  3. **DRL(Data Retrieval Language) Queries**
    * select
    


### <font color='#009900'>**1. DDL (Data Definition Language) Queries**</font>

* The purpose of the DDL queries is that "**To deal with the designing of the Tables (OR) Structure of the Tables**".
* We have three type of DDL queries, They are
    * **CREATE**
    * **ALTER**
    * **DROP**
---

* <font color='#009900'>**CREATE**</font>

    * **CREATE** command is used to creating a table.
    
<div class="alert alert-block alert-success">
<b>Syntax(Oracle):</b><br> 
<b>

</b>

    SQL>CREATE TABLE student(stno number(2) primary key, name varchar2(15) not null, marks number(5,2) not null );
    
<b>Syntax(Python):</b><br> 
<b>
    
    query="create table <table name>col1 datatype, col2 datatype,.... col-n datatype)" 
    
    cursor.execute(query)
</b> 
    
##### Example(Python):

    cursor.execute(cq)

### <font color='#009900'>**2. DRL (Data Retrival Language) Queries**</font>

* The purpose of DRL Query is that "**To Read the records from the table.**"
* To read the  records from the table , we use  **select** command.

<div class="alert alert-block alert-success">
<b>Syntax1(Oracle):</b><br> 
    <b><code>SQL>select col1, col2...col-n from tablename;</code>
</b>
</div>

<div class="alert alert-block alert-success">
<b>Syntax2(Oracle):</b><br> 
    <b><code>SQL>select col1, col2...col-n from tablename where cond list;</code>
</b>
</div>



* After executing the select query from python program , all the records are available in cursor object. To get the records from cursor object then we use 3 pre-defined Functions. They are
    * <font color='#009900'>**a) fetchone()**</font>
    * <font color='#009900'>**b) fetchmany(no. of records)**</font>
    * <font color='#009900'>**c) fetchall()**</font>
    
    
* a) <font color='#009900'>**fetchone()**</font> is used for obtaining only one record from cursor object where ever cusror object is pointing.
* b) <font color='#009900'>**fetchmany(no. of records)**</font> is used for obtaining sepecified number of records from cursor object. 
   here 

<div class="alert alert-block alert-info">
<b>Note:</b><br>
    i) if  we specify the "no.of records"  as <b>-ve value then we never get any output</b>.</br>
       ii) if  we specify the "no.of records"  as <b>+ve(more number of records than available) then we get all the records.</b></br>
       iii) if  we specify the "no.of records"  as  <b>0  then we get all the records.</b><br>
</div> 

   

* c) <font color='#009900'>**fetchall()**</font> is used obtaining all records from cursor object in the form list of tuples.
