<font color='#D2691E'>**Qn) write a python program which will insert employe number, ename and salary in a database in MySQL - Dynamic records**</font>

```python
# insertex2.py
# write a python program which will insert employe record dynamically in to MySQL DB by reading employee data from keyboard. - Dynamic records

import mysql.connector
import logging
import sys

logging.basicConfig(filename="insertex2.log", filemode='w', level=logging.INFO,
                    format='%(asctime)s %(levelname)s %(name)s %(message)s')


def insertrecord():
    while (True):
        try:
            logging.info("Trying to establish the MySQL DB connection")
            conn = mysql.connector.connect(host="localhost", user="root", password="root", database="testdb")
            logging.info("Successfully connected to MySQL DB")
            cur = conn.cursor()
            logging.info("Created the cursor object")
            logging.info("Accept Employee data dynamically from keyboard continuously")
            # Acaept the record Dynamically from the keyboard
            logging.info("Trying to insert the data into the employee table")
            eno = int(input("Enter employee number:"))
            ename = input("Enter employee name:")
            sal = float(input("Enter employee salary:"))
            iq = "insert into employee values(%d,'%s',%f)"
            cur.execute(iq % (eno, ename, sal))
            # (OR)   cur.execute("insert into employee values(%d, '%s', %f)" %(eno, ename, sal))
            logging.info("{} employee record inserted Successfully".format(cur.rowcount))
            conn.commit()
        except mysql.connector.DatabaseError as db:
            logging.exception(db)
        except TypeError as te:
            logging.exception(te)
        except ValueError:
            logging.exception("Dont enter str/symbol/alpha-numeric instead of number.")

        ch = input("Do you want to insert another record(yes/no):")
        if ch == "no":
            print("Thanks for using this program!...")
            sys.exit()


# main program
insertrecord()
```

<font color='#D2691E'>**Qn)write a python program which will delete a record based on employee number from MySQL DB**</font>

```python
# deleterecordex1.py
# write a python program which will delete a record based on employee number from MySQL DB

import mysql.connector
import logging
import sys

logging.basicConfig(filename="deleterecordex1.log", level=logging.INFO,
                    format='%(asctime)s %(levelname)s %(name)s %(message)s')


def empdelete():
    while (True):
        try:
            logging.info("Trying to establish the MySQL database connection.")
            conn = mysql.connector.connect(host="localhost", user="root", password="root", database="testdb")
            logging.info("Successfully established the database connection")
            cur = conn.cursor()
            logging.info("MySQL Database Cursor Object Created")
            logging.info("Accept the employee number form KBD")
            eno = int(input("Enter the employee number:"))
            # dq = "delete from employee where eno=1"
            logging.info("Trying to delete the record from the employee table")
            cur.execute("delete from employee where eno=%d" % (eno))
            conn.commit()
            if (cur.rowcount > 0):
                logging.info("Successfully deleted {} record from the employee table".format(cur.rowcount))
            else:
                logging.info("Record does not exist.")
        except mysql.connector.DatabaseError as dbe:
            logging.exception(dbe)
        except TypeError as te:
            logging.exception(te)
        except ValueError:
            logging.error("Don't enter str/Symbols/alpha-numeric in place of numbers")
        ch = input("Do you want to delete another record(yes/no):")
        if ch == "no":
            logging.info("Thanks for using this program!......")
            sys.exit()


# main program
empdelete()
```

```python
# updateex1.py
# write a python program which will update a record based on employee number in the MySQL DB

import mysql.connector
import logging
import sys

logging.basicConfig(filename="updateex1.log", level=logging.INFO,
                    format='%(asctime)s %(levelname)s %(name)s %(message)s')


def empupdate():
    while (True):
        try:
            logging.info("Trying to establish the MySQL database connection.")
            conn = mysql.connector.connect(host="localhost", user="root", password="root", database="testdb")
            logging.info("Successfully established the database connection")
            cur = conn.cursor()
            logging.info("MySQL Database Cursor Object Created")
            logging.info("Accept the employee number form KBD")
            eno = int(input("Enter the employee number:"))
            sal = float(input("Enter the salary of the employee:"))
            logging.info("Trying to update the record in the employee table")
            cur.execute("update employee set sal=%f where eno=%d" % (sal, eno))
            conn.commit()
            if (cur.rowcount > 0):
                logging.info("Successfully deleted {} record from the employee table".format(cur.rowcount))
            else:
                logging.info("Record does not exist.")
        except mysql.connector.DatabaseError as dbe:
            logging.exception(dbe)
        except TypeError as te:
            logging.exception(te)
        except ValueError:
            logging.error("Don't enter str/Symbols/alpha-numeric in place of numbers")
        ch = input("Do you want to delete another record(yes/no):")
        if ch == "no":
            logging.info("Thanks for using this program!......")
            sys.exit()


# main program
empupdate()
```

<font color='#D2691E'>**Qn)write a python program which will retrieve all the records of employee table whose salary is less than 7 lakhs**</font>

```python
# selectex1.py
# write a python program which will retrieve all the records of employee table whose salary is less than 7 lakhs

import mysql.connector
import logging
import sys

logging.basicConfig(filename="selectex1.log", filemode='a', level=logging.INFO,
                 format='%(asctime)s %(levelname)s %(name)s %(message)s')

# To redirect the print output to the log file

log = open("selectex1.log", 'a')
sys.stdout = log


def readrecord():
    try:
        logging.info("Trying to connect to the MySQL database")
        conn = mysql.connector.connect(host="localhost", user="root", password="root", database="testdb")
        logging.info("Successfully established the database connection")
        cur = conn.cursor()
        logging.info("MySQL Database Cursor Object Created")
        logging.info("Accept the employee number form KBD")
        # sq = "select * from employee where sal > 7"
        sq = "select * from employee where sal < 7"

        cur.execute(sq)
        print("-" * 50)
        print("\tE m p l o y e e  \tR e c o r d s")
        print("-" * 50)
        # To Display the column names
        columns = cur.description
        # print(columns[0][0],"\t",columns[1][0],"\t",columns[2][0])
        colnames = [str(var[0]) for var in columns]
        for colmname in colnames:
            print("\t{}".format(colmname), end="")
        print()
        print("-" * 50)
        # To display the records
        records = cur.fetchall()
        for record in records:
            for val in record:
                print("\t{}".format(val), end="")
            print()
        print("-" * 50)
        logging.info("Select query executed successfully")
    except mysql.connector.DatabaseError as de:
        logging.exception(de)
    except TypeError as te:
        logging.exception(te)
    except ValueError:
        logging.error("Don't enter str/Symbols/alpha-numeric in place of numbers")
    except NameError as ne:
        logging.exception(ne)


# main program
readrecord()
```
