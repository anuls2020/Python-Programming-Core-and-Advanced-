###  How to enable nbextensions in out Jupyter notebook?

Once we install the nbextension on our jupyter notebook using pip, use the following link to navigate through the nbextension page where we can see a variety of extensions we can use for developing more powerful and visually appealing Jupyter notebooks. 
http://localhost:8888/nbextensions

Once we were in this page,by default the configurable extensions will be disabled. First of all we need to uncheck the box and make them editable.



Once the extension boxes were editable, we can goahead and add the required packages in to our enveronments.
