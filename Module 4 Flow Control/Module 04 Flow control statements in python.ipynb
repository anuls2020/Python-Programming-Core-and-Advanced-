# Flow Control Statements in Python

* The purpose of Flow control statements in python is that "<font color='#009900'>To perform certain operation **one time** ( Perform X-Operation in the case of True (or) Peform Y-Operation in the case of False ) (or) **Peform certain operation repeatedly for finite number of times until Condition is False**.</font>"
* In Python Programming, we have 3 types of Flow control statements in python. They are

  1. <font color='#009900'><b>Conditional (or) Selection (or) Branching Statements</b></font>
  2. <font color='#009900'><b>Looping (or) Iterative (or) Repeatative Statements</b></font>
  3. <font color='#009900'><b>Misc Control statements.</b></font>

#### <font color='red'>d) match...case statement</font><br>

* match casse statement is similar to switch statement in other languages. 

**Syntax 1:-**

<code><font color='#D2691E'><b>match ChoiceExpression:
    case label-1: block of statement 1
    case label-2: block of statement 2
    .
    .
    case label-n: block of statement n
    case _:
        default Case Block statements
    Other Statements in the program</b></font></code>

**Explanation:**

* The ChoiceExpression can be either **int, str, bool** etc  (<font color='#009900'><b>except float and complex</b></font>)
* If the Value of ChoiceExpression is equal to Case Label1 then PVM executes corresponding Block of statements-1 and later executes other statements in Program.
* If the Value of ChoiceExpression is **not equal** to Case Label1 then PVM  compares Value of ChoiceExpression with  Case Label2  and if it is equal then executes corresponding Block of statements-2 and later executes other statements in Program.
* This process will be continued with all case labels. In general if the value of Choice Expression is equal to any of the specified Case Labels then PVM executes corresponding block of statements and **later executes Other statements in Program**.
* If the Value of ChoiceExpression **is not matching with any case labels** then PVM executes the block of statements written within <font color='#009900'><b>default case block (case _:)</b></font> and later exeutes Other statements in Program.

### <font color='#3498DB'>2) Looping (or) Iterative (or) Repeatative Statements</font>

* The purpose of Looping (or) Iterative (or) Repeatative Statements is that "**To perform Certain Operation repeatedly for finate number of times until condition is False**"
* In Python Programming, we have 2 types of looping statements. They are

     a) <font color='#009900'><b>while loop (OR) while .. else loop</b></font>
    
     b) <font color='#009900'><b>for loop (OR) for..else loop</b></font>

#### <font color='#3498DB'>a) while loop (OR) while .. else loop</font>

* At the time of dealing with looping statements, we must consider 3 points, They are,
   
     a) <font color='#009900'>Initilization Part</font><br>
     b) <font color='#009900'>Condition Part</font><br>
     c) <font color='#009900'>Updation Part</font><br>
  
**Syntax 1:-**
   
<code><font color='#D2691E'><b>while(Test Condition):
    statement-1
    statement-1
    .
    .
    statement-n
Other Statements in the program</b></font></code>
     
**Syntax 2:-**

<code><font color='#D2691E'><b>while(Test Condition):
    statement-1
    statement-1
    .
    .
    statement-n
else:
	else block of statements
    .
    .
    Other Statements in the program</b></font></code>



**Syntax Explenation:**

* Here 'while' and 'else' are the keywords
* Test condition result may be True of False
* In the while loop, if the test condition is true then PVM executes Indentation block of statements and once again PVM control goes to Test Cond. If the Test Cond is once again True then PVM executes Indentation block of statements once again. This Process will be continued until Test Cond becomes False.
* Once The test cond becomes False then PVM execute else block of statememnts, which are written in else block and later also executes other statements in program.

#### <font color='#3498DB'>b) for loop (OR) for..else loop</font>

**Syntax 1:-**
   
<code><font color='#D2691E'><b>
    for varname in Iterable_Object:
    statement-1
    statement-1
    .
    .
    statement-n
    ..........................
    Other Statements in the program</b></font></code>
     
   **OR**
     
**Syntax 2:-**

<code><font color='#D2691E'><b>
    for varname in Iterable_Object:
        statement-1
        statement-1
        .
        .
        ..........................
        statement-n
    else:
        else block of statements
        .
        .
        ..........................
    Other Statements in the program
</b></font></code>

**Explanation:**

* Here  '**for**' and '**in**' are the keywords
* The execution process of for loop is that "**Each element of Iterable-object kept in varname and executes Indentation Block of statements until all elements in iterable object are completed**"
* Here writing '**else**' block is **optional**.
* After for loop excution, condition becomes false and  PVM executes else block of statements(if we write else) and later executes Other statements in Program

<font color='#3498DB'><b>break statement</b></font>

* break is a key word
* The purpose of break statement is that "**To terminate the execution of loop logically when certain condition is satisfied  and PVM control comes of corresponding loop and executes other statements in the program**".

**Syntax 1:-**
   
<code><font color='#D2691E'><b>
    for varname in Iterable_Object:
        ..........................
        if (test cond):
            break
        ..........................
        ..........................
    Other Statements in the program
    </b></font></code>

**Syntax 2:-**
   
<code><font color='#D2691E'><b>
    while(Test Cond-1):
        ..........................
        if (test cond2):
            break
        ..........................
        ..........................
    Other Statements in the program
    </b></font></code>


<font color='#3498DB'><b>Continue statement</b></font>

* continue is a keyword
* continue statement is used for **making the PVM to go to the top of the loop without executing the following statements which are written after continue statement for that current Iteration only**.
* continue statement  to be **used always inside of loops**.
* when we use continue statement **else part of corresponding loop also executes provided loop condition becomes false**.


**Syntax:-**
   
<code><font color='#D2691E'><b>
    for varname   in Iterable-object:
        ..........................
        if (test cond2):
            continue
        statement-1    # written after continue statement
        statement-2
        statement-n
        ..........................
    Other Statements in the program
    </b></font></code>
