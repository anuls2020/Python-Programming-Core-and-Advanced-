### Syntax for defining a class

* We know that every Program of Python by using OOPs must starts with the concept of classes.
* The Syntax for defining a class is shown bellow

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
</div>

```python    
class  <clsname>:
   Class  Level Data Members
   def  instancemethodname(self,list of formal params if any ):   <------- No decorator for instance_methods
      -------------------------------
      Block of stattaments--specific Operations
      Specify Instance Data members
      --------------------------------
   @classmethod                                                   <-------- Dcorator                   
   def  classlevelmethodname(cls,list of formal params if any ):
     --------------------------------
     Block of statements--class level operations
     Specify Class  Level Data Members
     --------------------------------
   @staticmethod                                                  <-------- Decorator
   def  staticmethodname(list of formal params if any):           <-------- Only Formal Parameters
     --------------------------------
     Block of statements--Utility Operations
     --------------------------------
```

#### Diagramatic representation of Instance Level and Class Level Data Members

#### <font color='#009900'>1. Instance Data Members</font>

* The Instance Data Members are those whose <font color='#3498DB'>**memory space is created each and every time when an object is created**</font>. 
* The Instance Data Members are <font color='#3498DB'>**used for storing specific values OR particular values**</font> which are suitable for one person or programmer.
* The instance data members must be <font color='#3498DB'>**ACCESSED or PROCESSED**</font> with respect to <font color='#3498DB'>**Object_Name OR self**</font>. 

<div class="alert alert-block alert-info">
<b>Tip/Note:self is an Implicit Object</b>
  
  - Along with the object what we have created, one more object is created internally, which is called <font color='#3498DB'>**"self"**</font>, it will carry the **address/id** of that object. 
  - Whenever we call a new object the "self" object take the new address to point the newly created object. 
  - "self" is one of the **Pre Defined Attribute**
  - "self" is **anattribute which is pointing to an object internally**, Therefore it is called **implicit object**. 
</div>

#### <font color='#3498DB'>**Instance Methods:**</font>

     
* Instance methods are used for performing specific operations and hence the <font color='#009900'>**Instance Methods**</font> are also called "<font color='#009900'>**Object Level Methods**</font>"
* Programatically, Instance Methods always takes "<font color='#009900'>**self**</font>" as First Formal Parameter for <font color='#009900'>**storing address(id) / reference of object**</font.

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
</div>

```python    
def  instancemethodname(self,list of formal params if any ):   <------- No decorator for instance_methods
     -------------------------------
     Block of stattaments--specific Operations
     Specify Instance Data members
     --------------------------------
```

* In Python Programming, all Instance Methods <font color='#009900'>**MUST BE ACCESSED w.r.t Object Name (or) self**</font>.

```python
    ObjectName.Instance Method Name()
      (OR)
    self.Instance Method Name()
```

* Here <font color='#009900'>**self**</font> is a shadow copy of the object it is being called. Both carry the same address. See the below program.
* self is called <font color='#009900'>**inplicit object**</font> for current class object

**Example 4**

````python
# studentex4.py
# This program will store stno, sname, marks and course with OOPS by using methods

class Student:
    def getstudentdetails(self):  # Here self is called inplicit object for current class object
        print("id of self in getstudentdetails():", id(self))


# main program
s1 = Student()
print("id of s1:", id(s1))
s1.getstudentdetails()
print("--------------------------------------------------")
s2 = Student()
print("id of s2:", id(s2))
s2.getstudentdetails()
````

#### <font color='#3498DB'>**Class Level Methods**</font>

* Class Level Methods are used for performing <font color='#009900'>**Common Operations on Class Level Data Members**</font> and hence they named as class level methods, We can define/specify <font color='#009900'>**Class Level Data Members**</font> inside the Class Level Methods. 
* The Class level methods must be preceded with a <font color='#009900'>**pre-defined decorator**</font> called <font color='#009900'>**@classmethod**</font> and it must take first formal parameter as "<font color='#009900'>**cls**</font>" 

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
</div> 

```python 
@classmethod                     
def  classlevelmethodname(cls,list of formal params if any ):
     -------------------------------
     Block of stattaments--class level operations
     Specify Class Data members
     --------------------------------
```

* In Python programming, all Class Level Methods must be accessed with respect to "<font color='#009900'>**Class_Name**</font>" (OR) "<font color='#009900'>**Object_Name**</font>" (OR) "<font color='#009900'>**cls**</font>" (OR) "<font color='#009900'>**self**</font>".
 
 ```python
    className.ClassLevelMethodName()
        OR
    Object_Name.ClassLevelMethodName()
        OR
    cls.ClassLevelMethodName()
        OR
    self.ClassLevelMethodName()
```

**Example 1: Demonstration of class level methos with single object calling**

````python
# employeeex1.py
# This program demonstrate the specification of Class Level Data members

class employee:
    @classmethod
    def getcompname(cls):  # Class level Method
        cls.cname = "Scotiabank"

    @classmethod
    def getcompanyadd(cls):  # Class level Method
        employee.city = "Toronto"

    def getempdet(self):
        self.eno = int(input("Enter the Employee Number:"))
        self.ename = input("Enter the Employee Name:")

    def dispempdet(self):
        print("-" * 50)
        print("Employee Number: {}".format(self.eno))
        print("Employee Name: {}".format(self.ename))
        print("Employee Company Name: {}".format(employee.cname))
        print("Employee Company Address: {}".format(employee.city))
        print("-" * 50)


# main program

eo1 = employee()
eo1.getempdet()  # Calling instance method
employee.getcompname() # Calling class level method
employee.getcompanyadd() # Calling class level method
eo1.dispempdet()  # Calling instance method
````

**Example 2: Demonstration of class level methos with multiple object calling**

````python
# employeeex2.py
# This program demonstrate the specification of Class Level Data members

class employee:
    @classmethod
    def getcompname(cls):  # Class level Method
        cls.cname = "Scotiabank"

    @classmethod
    def getcompanyadd(cls):  # Class level Method
        employee.city = "HYD"

    def getempdet(self):
        self.eno = int(input("Enter the Employee Number:"))
        self.ename = input("Enter the Employee Name:")

    def dispempdet(self):
        print("-" * 50)
        print("Employee Number: {}".format(self.eno))
        print("Employee Name: {}".format(self.ename))
        print("Employee Company Name: {}".format(employee.cname))
        print("Employee Company Address: {}".format(employee.city))
        print("-" * 50)


# main program
employee.getcompname() # Calling class level method
employee.getcompanyadd() # Calling class level method
eo1 = employee()
eo2 = employee()
eo1.getempdet()  # Calling instance method
eo2.getempdet()  # Calling instance method
eo1.dispempdet()  # Calling instance method
eo2.dispempdet()  # Calling instance method
````

#### <font color='#3498DB'>**Static methods**</font>

* Static Methods are used for performing <font color='#009900'>**Universal / Utility Operations**</font>.
* The Static Methods Definition must preceded with a **pre-defined decorator** called <font color='#009900'>**@staticmethod**</font> and it <font color='red'>**never takes** the First Formal Parameter as "**cls or self".**</font>

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
</div>

```python
@staticmethod 
def  staticmethodname(list of formal params if any ):
    --------------------------------
    Block of statements--Utility / Universal operations
    --------------------------------
    --------------------------------
 ```

* In Python Programming, all **Static Methods must be accessed** w.r.t <font color='#009900'>**ClassName**</font> (or) <font color='#009900'>**cls**</font> (or) <font color='#009900'>**Object Name**</font> (or) <font color='#009900'>**self**</font>.

**Example 1:** Static method ex1

````python
# This program will demonstrate the static method
# staticmethodex1.py

class Employee:
    def getempdet(self):
        self.eno = int(input("Enter the employee number:"))
        self.ename = input("Enter the employee Name:")


class Student():
    def getstuddet(self):
        self.stno = int(input("Enter the Student Number:"))
        self.sname = input("Enter the Student Name:")
        self.scname = input("Enter the College name:")


# Let's develop a class name with a Method to display the content of any object (Utility Method/Universal Operation)

class ObjectDisplay():
    @staticmethod
    def displaydata(obj):
        print("Type of Obj", type(obj))

# main program
eo = Employee()
eo.getempdet()

so = Student()
so.getstuddet()

# Calling Static Method of ObjectDisplay class
ObjectDisplay.displaydata(eo)
ObjectDisplay.displaydata(so)
````

**Example 3:** staticmethodex3 Display only the contents **with respect class name**

````python
# This program will demonstrate the static method
# staticmethodex3.py

class Employee:
    def getempdet(self):
        self.eno = int(input("Enter the employee number:"))
        self.ename = input("Enter the employee Name:")


class Student():
    def getstuddet(self):
        self.stno = int(input("Enter the Student Number:"))
        self.sname = input("Enter the Student Name:")
        self.scname = input("Enter the College name:")


# Let's develop a class name with a Method to display the content of any object (Utility Method/Universal Operation)

class ObjectDisplay():
    @staticmethod
    def displaydata(obj):
        # print("Type of Obj", type(obj))
        # print("Content of obj:", obj.__dict__)
        print("-" * 50)
        for k, v in obj.__dict__.items():
            print("{} --> {}".format(k, v))
        print("-" * 50)


# main program
eo = Employee()
eo.getempdet()

so = Student()
so.getstuddet()

# Calling Static Method of ObjectDisplay class
print("Employee details")
ObjectDisplay.displaydata(eo)  # Calling static method w.r.t corresponding class name
print("Student details")
ObjectDisplay.displaydata(so)
````

## <font color='#3498DB'>**Objects**</font>

#### Importance of Object:

* When we define a **class , Memory space is not created for Data Members and Methods** But whose **Memory Space is created when e create an object.**
* When we define a class and **class name can be treated as Data Type** but we can't store the data. **To Store the data , we must create an object w.r.t Class Name.**
* To **do any data processing**, we **must create an object.**
* **To create an object**, there **must exist a class definition** otherwise we get error.

**Demonstration of the OOPS**

````python
# This program will find the sum of two numbers using OOPs
# sum.py

class Sum:
    def getvalues(self):
        self.a = float(input("Enter the first value: "))
        self.b = float(input("Enter the second value: "))

    def findsum(self):
        self.c = self.a + self.b

    def dispresult(self):
        print("sum of {} and {} is {} ".format(self.a, self.b, self.c))

# main program
so=Sum()
print("content of so before method calling: ", so.__dict__)
so.getvalues()
print("content of so after method calling: ", so.__dict__)
so.findsum()
print("content of so after sum is over: ", so.__dict__)
so.dispresult()
````
