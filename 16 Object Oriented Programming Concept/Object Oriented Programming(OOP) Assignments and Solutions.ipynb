```python
# This python program determines the result of relational operators by accepting two values form keyboard with classes and objects.
# arith.py

class Values:
    def getvalues(self):
        try:
            self.a = float(input("Enter the first value: "))
            self.b = float(input("Enter the second value: "))
            self.op = input("Enter any relational operator: ")
        except ValueError:
            print("Dont enter string/symbol/alpha-numeric for number")


class Calculator:
    @staticmethod
    def relateional(obj):
        try:
            match (obj.op):
                case ">":
                    print("{} is greater than {} ==> {}".format(obj.a, obj.b, obj.a > obj.b))
                case "<":
                    print("{} is less than {} => {}".format(obj.a, obj.b, obj.a < obj.b))
                case "=":
                    print("{} is equal to {} => {}".format(obj.a, obj.b, obj.a == obj.b))
                case "!=":
                    print("{} is not equal to {} => {}".format(obj.a, obj.b, obj.a != obj.b))
                case "<=":
                    print("{} is less than {} => {}".format(obj.a, obj.b, obj.a // obj.b))
                case "<=":
                    print("{} is less than {} => {}".format(obj.a, obj.b, obj.a % obj.b))
                case _:
                    print("{} is not an arithmetic operator".format(obj.op))
        except AttributeError:
            print("Arithmetic Operator missing for the operations")
        except ZeroDivisionError:
            print("For division, the denominator should not be zero")


# main program
vo = Values()
vo.getvalues()
Calculator.relateional(vo)
```

```python
"""Write a python program which will calculate roots of quadratic equation by using classes and objects"""
# QuardEqn.py

import cmath

class Quardeqn:
    def getvalues(self):
        self.a=float(input("Enter the value of a:"))
        self.b=float(input("Enter the value of b:"))
        self.c=float(input("Enter the value of c:"))

    def discriminant(self):
        self.discr = self.b**2 - 4*self.a*self.c

    def quardroots(self):
        self.sol1 = (-self.b + cmath.sqrt(self.discr))/(2*self.a)
        self.sol2 = (-self.b - cmath.sqrt(self.discr))/(2*self.a)
        print("The roots are {} and {}".format(self.sol1, self.sol2))

#main program
qo = Quardeqn()
qo.getvalues()
qo.discriminant()
qo.quardroots()
```

```python
# MethodOverridingEx3.py
# Program for calculating the area of figures such as Circle, Rectangle and Square using Method Overriding

class Circle:
    def area(self):
        self.r = float(input("Enter the radius of the circle: "))
        self.ac = 3.14 * self.r ** 2
        print("Area of the Circle: {}".format(self.ac))
        print("-" * 40)


class Square(Circle):
    def area(self):
        self.a = float(input("Enter the length of the square: "))
        self.sa = self.a ** 2
        print("Area of the Square: {}".format(self.sa))
        print("-" * 40)
        super().area()


class Rectangle(Square):
    def area(self):
        self.l, self.b = float(input("Enter the length: ")), float(input("Enter the bredth: "))
        self.ar = self.l * self.b
        print("Area of the Rectangle: {}".format(self.ar))
        print("-" * 40)
        super().area()


# main program

ro = Rectangle()
ro.area()
```

```python
# This Program will demonstrate Hybrid inheritance by using Method Overriding for implementing Polymorphysm
# HybridInh1.py

class C1:
    def fun(self):
        print("This is from fun() --> C1")


class C2(C1):
    def fun(self):
        print("This is from fun() --> C2")


class C3(C1):
    def fun(self):
        print("This is from fun() --> C3")


class C4(C2, C3):
    def fun(self):
        print("This is from fun() --> C4")


class C5(C4):
    def fun(self):
        print("This is from fun() --> C5")


class C6(C4):
    def fun(self):
        print("This is from fun() --> C6")


class C7(C5, C6):
    def fun(self):
        print("This is from fun() --> C7")
        C5.fun(self)
        C6.fun(self)
        C4.fun(self)
        C3.fun(self)
        C2.fun(self)
        C1.fun(self)


# main program

O7 = C7()
O7.fun()
```

```python
# This Program will demonstrate Hybrid inheritance by using Method Overriding for implementing Polymorphism using constructors
# HybridInh2.py

class C1:
    def __init__(self):
        print("This is from Class C1")


class C2(C1):
    def __init__(self):
        print("This is from Class C2")


class C3(C1):
    def __init__(self):
        print("This is from Class C3")


class C4(C2, C3):
    def __init__(self):
        print("This is from Class C4")


class C5(C4):
    def __init__(self):
        print("This is from Class C5")


class C6(C4):
    def __init__(self):
        print("This is from Class C6")


class C7(C5, C6):
    def __init__(self):
        print("This is from Class C7")
        C5.__init__(self)
        C6.__init__(self)
        C4.__init__(self)
        C3.__init__(self)
        C2.__init__(self)
        C1.__init__(self)


# main program

O7 = C7()
```
