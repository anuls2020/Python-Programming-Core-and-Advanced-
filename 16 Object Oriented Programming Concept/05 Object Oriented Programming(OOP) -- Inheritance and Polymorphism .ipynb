### Introduction to Inheritance  


* Inheritance is one of the distinct principle of OOPs
* The purpose of Inheritance is that "**To Build / Develop Re-Usable Applications.**"
* The advantages of Inheritance concept is that
  - **Application Development time Less**
  - **Application Memory Space is Very Less**
  - **Application Execution Time is Very Less**
  - **Application Performnace is Enhanced (Improved)**
  - **Redundency ( Duplication ) of the code is minimized**.

### Definition of Inheritance:

* <font color='#3498DB'>**The process of obtaining the Data Members , Methods and Constructors (Features) from One class into another class is called Inheritance.**</font>
* The class which is giving the Data Memebrs , Methods and Constructors (Features) is called "<font color='#3498DB'>**Base / Super / Parent Class.**</font>"
* The class which is Taking the Data Memebrs , Methods and Constructors (Features) is called "<font color='#3498DB'>**Derived / Sub / Child Class.**</font>"
* The Inheritance Priciple <font color='#3498DB'>**always Provides Logical Memory Management. This Memory management says that Neither we write Physical Source Code Nor takes Physical Memory Space.**</font>

### Types of Inheritances

<font color='#3498DB'>**Syntax for implementing Inheritance**</font>

Inheriting the features of Base Class into Derived Class
* The features of Base Class are nothing but Data Members , Methods and Constructors.
* To Inherit the features of Base Class into Derived Class, we use the following Syntax:

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
</div>

```python
class  <class-name-1>:    # Base Class
		 ---------------------
		---------------------
class  <class-name-2>:    # Base Class
		 ---------------------
		---------------------
class  <class-name-n>:    # Base Class
		 ---------------------
		---------------------
class  <class-name-n+1>( Class-Name-1,Class-Name-2...Class-Name-n):   
		 ---------------------
		---------------------
```

**Explanation:**

* Here ```<class-name-1>, <class-name-2>....<class-name-n>``` are called <font color='#3498DB'>**Base Classes.**</font>
* ```<class-name-n+1>``` is called <font color='#3498DB'>**Derived Class.**</font>
* When we develop any Inheritance Based Application, <font color='#3498DB'>**It is always recommended to create an object of Bottom most Derived Class bcoz It contains all the features of Intermediate Base Classes and Top Most Base Class.**</font>
* For Every class in python, <font color='#3498DB'>**There exists an implicit pre-defined super class called**</font>  <font color='#009900'>**"object"**</font> and it provdes <font color='#009900'>**"Garbage Collector Program"**</font>


**Simple inheritance Example**

````python
# This program demonstrate the concept of Inheritance

class C1:
    def setC1Values(self):
        self.id = 10
        self.crs = "Python"


class C2(C1):  # Following single inheritance
    def setC2Values(self):
        self.sname = "Rossum"
        self.cname = "MIT"


# main Program

o2 = C2()
print("Content of o2 before setting values:", o2.__dict__)
o2.setC1Values()  # w.r.t Derived Class, we are calling Base Class Methods
o2.setC2Values()
print("Contents of o2 after setting values:", o2.__dict__)
````

python program for implementing the following

1. Let's assume there exist a university which contains university name and location. aaccept and display university details

2. Let's assume there exist a college which contains college name, and college location. Accept and display college details along with university details.**Single Inheritance**

3. Let's assume there exist a student which contains student number, , student name, and course persuing. Accept and display student details along with college and university details.**Multi Level Inheritance**

````python
# InhProgEx4.py
"""
1. Let's assume there exist a university which contains university name and location. accept and
display university details
2. Let's assume there exist a college which contains college name, and college location. Accept and display college
details along with university details.
3. Let's assume there exist a student which contains student number, , student name, and course persuing. Accept and
display student details along with college and university details

"""


class University:
    def setunivalues(self):
        self.uname = input("Enter the University name: ")
        self.uloc = input("Enter the University location: ")

    def dispunivalue(self):
        # print("-" * 40)
        print("University name: {}".format(self.uname))
        print("University location: {}".format(self.uloc))
        # print("-" * 40)


class College(University):
    def setcoldetails(self):
        self.cname = input("enter the college name: ")
        self.cloc = input("enter the college location: ")
        self.setunivalues()

    def dispcolldetails(self):
        # print("-" * 40)
        print("College Name: {}".format(self.cname))
        print("College location: {}".format(self.cloc))
        self.dispunivalue()
        # print("-" * 40)


class Student(College):
    def setstuddet(self):
        self.sno = int(input("Enter student number: "))
        self.name = input("Enter student name: ")
        self.crs = input("Enter course name: ")
        self.setcoldetails()

    def dispstudetails(self):
        print("-" * 40)
        print("Student Number: {}".format(self.sno))
        print("Student Name: {}".format(self.name))
        print("Student Course: {}".format(self.crs))
        self.dispcolldetails()
        print("-" * 40)


# main Program

so = Student() # Create an object of bottom most derived class Employee
so.setstuddet()
so.dispstudetails()
````

## <font color='#3498DB'>**Polymorphism**</font>

* <font color='#3498DB'>**Polymorphism**</font> is one of the distinct principle of OOPs
* The purpose of Polymorphism principle is that "<font color='#3498DB'>**To build the Object oriented Applications with effective Memory space(less memory space)".**</font>

* Deffinition of Polymorphism:

* The process of representing "<font color='#3498DB'>**One Form in Multiple Forms**</font>" is called Polymorphism.
* In the definition of Polymorphism, <font color='#3498DB'>**One Form represents Original Method of Base Class and Multiple Forms represents Overridden Methods of Derived Class.**</font>
* To <font color='#009900'>**Implement Polymorphism Principle, we must use Method Overriding.**</font>

### <font color='#3498DB'>**Number approaches to call Base class methods from Derived Class Methods**</font>

* In Python Programming, we have two approaches to calle base class methods from derived class methods. They are
  - <font color='#009900'>**a) By Using  super()**</font>
  - <font color='#009900'>**b) By using  ClassName**</font>

<font color='#3498DB'>**a) By Using  super():**</font>

* **super()** is used for calling Base Class Orginal Method (or) Original Constructor from derived class Overridden Method / Constructor.

<div class="alert alert-block alert-success">
<b>Syntax 1:</b><br> 
</div>

```python
super().MethodName()
super().MethodName(list of values)
```

<div class="alert alert-block alert-success">
<b>Syntax 2:</b><br> 
</div>

```python
super().__init__()
super().__init__(list of values)
```
   
				
* With <font color='#3498DB'>**super()**</font> we are <font color='#3498DB'>**able to call single base class method from derived class method**</font> but <font color='red'>**unable to call multiple base class methods from derived class methods.**</fot>
* <font color='#3498DB'>**To Overcome this problem**,</font> we must use <font color='#009900'>**Class Name concept.**</font>

<font color='#3498DB'>**b) By using  ClassName:**</font>

* <font color='#009900'>**By using Class Name concept we can call  Multiple Base Class Orginal Methods (or) Original Constructors from derived class Overridden Method / Constructor.**</font>

<div class="alert alert-block alert-success">
<b>Syntax 1:</b><br> 
</div>

```python
Class Name.MethodName(self)
Class Name.MethodName(self , list of values)
```

<div class="alert alert-block alert-success">
<b>Syntax 2:</b><br> 
</div>

```python
Class Name.__init__(self)
Class Name.__init__(self , list of values)
```

````python
# MethodOverridingEx3.py
# This program is purely using Inheritance with method overriding
class Circle:
    def draw(self):  # Original Method
        print("Drawing Circle!.. ")
        # super().draw() # -- Error : AttributeError: 'super' object has no attribute 'draw'


class Rectangle(Circle):
    def draw(self):  # Overridden Method
        print("Drawing Rectangle!..")
        super().draw()  # Calling Base Class Method from Derived Class.


class Square(Rectangle):
    def draw(self):
        print("Drawing Square!..")
        super().draw()  # Calling Intermediate Base Class Method from Derived Class.


# main program

so = Square()
so.draw()
````

````python
# MethodOverridingEx4.py
# Program for calculating the area of figures such as Circle, Rectangle and Square using Method Overriding

class Circle:
    def area(self): # Original Method -- One form
        self.r = float(input("Enter the radius of the circle: "))
        self.ac = 3.14 * self.r ** 2
        print("Area of the Circle: {}".format(self.ac))
        print("-" * 40)


class Square(Circle):
    def area(self):
        self.a = float(input("Enter the length of the square: "))
        self.sa = self.a ** 2
        print("Area of the Square: {}".format(self.sa))
        print("-" * 40)
        super().area()


class Rectangle(Square):
    def area(self):
        self.l, self.b = float(input("Enter the length: ")), float(input("Enter the bredth: "))
        self.ar = self.l * self.b
        print("Area of the Rectangle: {}".format(self.ar))
        print("-" * 40)
        super().area()


# main program

ro = Rectangle()
ro.area()
````

````python
# This Program will demonstrate Hybrid inheritance by using Method Overriding for implementing Polymorphysm
# HybridInh1.py

class C1:
    def fun(self):
        print("This is from fun() --> C1")


class C2(C1):
    def fun(self):
        print("This is from fun() --> C2")


class C3(C1):
    def fun(self):
        print("This is from fun() --> C3")


class C4(C2, C3):
    def fun(self):
        print("This is from fun() --> C4")


class C5(C4):
    def fun(self):
        print("This is from fun() --> C5")


class C6(C4):
    def fun(self):
        print("This is from fun() --> C6")


class C7(C5, C6):
    def fun(self):
        print("This is from fun() --> C7")
        C5.fun(self)
        C6.fun(self)
        C4.fun(self)
        C3.fun(self)
        C2.fun(self)
        C1.fun(self)


# main program

O7 = C7()
O7.fun()
````

````python
# This Program will demonstrate Hybrid inheritance by using Method Overriding for implementing Polymorphism using constructors
# HybridInh2.py

class C1:
    def __init__(self):
        print("This is from Class C1")


class C2(C1):
    def __init__(self):
        print("This is from Class C2")


class C3(C1):
    def __init__(self):
        print("This is from Class C3")


class C4(C2, C3):
    def __init__(self):
        print("This is from Class C4")


class C5(C4):
    def __init__(self):
        print("This is from Class C5")


class C6(C4):
    def __init__(self):
        print("This is from Class C6")


class C7(C5, C6):
    def __init__(self):
        print("This is from Class C7")
        C5.__init__(self)
        C6.__init__(self)
        C4.__init__(self)
        C3.__init__(self)
        C2.__init__(self)
        C1.__init__(self)


# main program

O7 = C7()
````
