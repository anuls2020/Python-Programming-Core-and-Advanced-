<h1><center>Packages in Python</center></h1> 


### <font color='#3498DB'>**Introduction**</font>


* We know that **FUNCTIONS** concept is used for <font color='#3498DB'>**Performing Certain operation and provides Code Re-usability within the program**</font> but <font color='red'>**not able to provide Code Re-usability across the programs**</font>.
* We know that **MODULES** concept is used for <font color='#3498DB'>**re-using the code across the programs provided the modules must present in same Folder**</font> but <font color='red'>**not able to get Code Re-usability across Folders / Drives / Environments / Networks..etc**</font>
* The **PACKAGES concept** is used for getting the <font color='#3498DB'>**Code Re-usability across Folders / Drives / Environments / Networks**..etc through modules where modules contains Variables, Functions and Classes.</font>

### <font color='#3498DB'>**Definition of Packages**</font>

<div class="alert alert-block alert-success">
<b>Definition:</b><br> <b> A Package is a collection of Modules.</b>
</div>


### <font color='#3498DB'>**General Sructure of Packages**</font>




### <font color='#3498DB'>**Strecture of a Package with Subpackages and imports**</font>

The attributes(variables, functions, modules etc..) used in one sub package can be re used in another subpackage by importing them

### <font color='#3498DB'>**Detailed Structure of a Package with multiple subpackages**</font>

### <font color='#3498DB'>**Step for developing the Packages**</font>

* Step-1: **Create a folder**
* Step-2: Define / place an an **empty python** file on the name of  `__init__.py` in folder to make the folder name as Package Name.
* Step-3: **Define / place a module (s) in the package(Folder Name).**

### <font color='#3498DB'>**Number of approaches to re-use the modules of Packages**</font>


* We have two approaches re-use the modules of Packages. They are

   1. By using <font color='#009900'>**sys.path.append()**</font>
   2. By using <font color='#009900'>**PYTHONPATH Environmental Variable**</font>
   

#### 1.   By using <font color='#009900'>**sys.path.append():**</font> 

<div class="alert alert-block alert-success">
<b>Syntax:</b><br> 
    <code><b>sys.path.append("Absolute path of Package Name")</b></code>

(Absolute path can be given either using back slash(`\`) or using Forward slash(`/`) )

    
Examples:

`#kvr1.py
import sys
sys.path.append("E:\KVR-PYTHON-7AM\PACKAGES\BANK")
from formula import simpleint
simpleint()`

(OR)
    
`#kvr1.py

#### 2.   By using <font color='#009900'>**PYTHONPATH Environmental Variable:**</font> 

<div class="alert alert-block alert-success">
