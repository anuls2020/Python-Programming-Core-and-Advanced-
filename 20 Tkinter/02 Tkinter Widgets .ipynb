# Tkinter Widgets

* <b>What is ment by Tkinter Widgets</b>?

 * Tkinter provides rich set of GUI elements such as <b>buttons, menu and other types of entry fields and display areas</b>.These are called <b>widgets</b>
 * Widget is an element of <b>Graphical User Interface (GUI) that displays/illustrates information or gives a way for the user to interact with the OS</b>.  
 * In Tkinter , <b>Widgets are objects; instances of classes that represent buttons, frames, and so on</b>.
 * Each <b>Widget is a Python Object</b>. Therefore, whenever we are creating a widget, we must have to pass it's parent as a parameter to the widget creation function which we are calling. 
 * There is an <b>exception for "root" window, which is the top-level window that will contain everything else and it doesn't have a parent</b>.

```python
# TkinterEx5.py
# This is to create the tkinter window
# add title to the applications, # Add geometry, # Add Widgets using ttk
import tkinter as tk
from tkinter import ttk

# Create a window
window = tk.Tk()

# Add title to the window
window.title("First window app!..")

# Add geometry
window.geometry('500x350')

# Create Widgets(tk widgets)
text = tk.Text(master=window)
text.pack()

# ttk widgets(ttk label)
label = ttk.Label(master=window, text='This is a text')
label.pack()

# Run
window.mainloop()
```

```python
# TkinterEx2.py
# This is to create the tkinter window
# add title to the applications, # Add geometry, # Add Widgets using ttk,
# Add entry box
import tkinter as tk
from tkinter import ttk


def button_func():
    print('A button was pressed')


# Create a window
window = tk.Tk()

# Add title to the window
window.title("First window app!..")

# Add geometry
window.geometry('800x700')

# Create Widgets(tk widgets)
text = tk.Text(master=window)
text.pack()

# ttk widgets(ttk label)
label = ttk.Label(master=window, text='This is a text')
label.pack()

# ttk entry
entry = ttk.Entry(master=window)
entry.pack()

# ttk button
button = ttk.Button(master=window, text='Enter', command=button_func)
button.pack()

# Run
window.mainloop()
```
