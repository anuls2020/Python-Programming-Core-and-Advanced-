### <font color='#3498DB'>**Pyrpose of Exception Handling**</font>

* The purpose of Exception handling is to <font color='#009900'>**Develop Robust(Strong) Appliations**</font>. 

* In Real time, To develop any project, we need to choose a Language.  And by using that language we can **develop**, **Compile** & **Execute** various programs. 
* During this process we may get **3-Types** of errors. They are,
    1. <font color='#009900'>**Compile Time Error**</font>
    2. <font color='#009900'>**Logical Error**</font>
    3. <font color='#009900'>**Runtime Error**</font>
    
#####  <font color='#3498DB'>**1. Compile Time Error**</font>
   * Compile time errors are those, which are occurring at compile time(**.py  --> .pyc**)
   * Compile time errors are occurring due to <font color='#009900'>**Syntaxes are not followed**</font> by the programmer.
   * These errors are solved by the Programmers at development level.  
   
   
##### <font color='#3498DB'>**2. Logical Error**</font>
   * Logical errors are those which are <font color='#009900'>**occurring during execution time**</font>. 
    

   * Logical Errors are occurring due to <font color='#009900'>**wrong representation of Logic**</font>.
   * Logical Errors always give **wrong results/inconsistant results**
   * These errors solved by Programmer during Development Time.


##### <font color='#3498DB'>**3. Runtime Error**</font>
   * Runtime errors are those, which are **occurring at execution time**.
   * Runtime errors are occurring due to <font color='#009900'>**wrong input or invalid input entered by application user or end users**</font>. 
   * Runtime errors must be **handled by programmer during development time with forcasting knowledge**.

### <font color='#3498DB'>**Points to be remembered in Exception Handling**</font>

* When the application or end user enters a wrong input/invalid input, then we get **Runtime Error**
* Runtime errors by default given **Technical Error Messages**. These messages are understandable by Programmers but not by the end users. Industry always recommends, convert the **Technical Error Messages** in to **User friendly Error Messages** by using <font color='#009900'>**"Excepotion Handling"**</font>. 
* Every exception by default gives **Technical Error Message**.

### <font color='#3498DB'>What happen internally when the exception occurs</font>

* When exception occurs in Python Program, Three steps take place internally.They are

  a. PVM **terminate the program execution** abnormally.</br>
  b. PVM **comes out of the program flow** without executing rest of the statements.</br>
  c. PVM by default **generates Technical Error Message**.</br>

* To do the steps a,b and c, PVM create an object of appropriate exception classes.   
* Hence every **Invalid Input** given **Exception** and **Every Exception** is treated as **Object** and is created w.r.t appropriate exception classes.

   <font color='#009900'>**Invalid Input --> Exception --> Object --> Appropriate Exception Classes**</font>

### <font color='#3498DB'>Types of Exceptions</font>

* In Python, We have two type of Exceptions,They are,
    1. <font color='#009900'>**Pre-Defined / Built-in Exceptions**</font>
    2. <font color='#009900'>**Programmer / User / Custom-defined Exceptions**</font>

#### <font color='#3498DB'>**Pre-Defined / Built-in Exceptions**</font>

* Pre-defined / Built-in Exceptions are developed by Python Language Developers and they are available in Python API(Library) and whos role is to deal with Universal problems. 
* Someof the Universal Problems are
  * a) Division by Zero Problems ( <font color='#009900'>**ZeroDivisionError**</font>)
  * b) Invalid number formats ( <font color='#009900'>**ValueError**</font>)
  * c) Invalid arguments / Operations (<font color='#009900'>**TypeError**</font>)
  * d) searching the value of value in in dict by passing invalid key  (<font color='#009900'>**KeyError**</font>)
  * e) Invalid Variable Names (<font color='#009900'>**NameError**</font>)......etc

#### <font color='#3498DB'>**Programmer / User / Custom-defined Exceptions**</font>

* Programmer / User / Custom-defined Exceptions  are developed by the programmer and are available in Python Project. 
* These exceptions were used to deal with common Problems,They are,
  * a) Attempting to enter invalid PIN in atm applicatioins.
  * b) Attempting to enter Wrong User Name and Password.
  * c) Attempting to withdraw Invalid amount from existing account
  * ...etc
  
* Every **Valid Input** given **Successful Output**
* Every **Invalid input** given **Exceptions**

### <font color='#3498DB'>**Handling the Exceptions in Python**</font>

* Handling Exception means converting "**Technical Error Message**" into "**User Friendly Error Message**"
* To convert  "**Technical Error Message**" into "**User Friendly Error Message**", we have 5 Keywords. They are, 
    1. <font color='#009900'>**try**</font>
    2. <font color='#009900'>**except**</font>
    3. <font color='#009900'>**else**</font>
    4. <font color='#009900'>**finally**</font>
    5. <font color='#009900'>**raise**</font>

### <font color='#3498DB'>**Explanation for the keywords used in Syntax for handling the exceptions**</font>

1. <font color='#009900'>**try block:**</font>

 * It is the block, in which we write block of statements generating exceptions. In otherwords what are all the statements are generating exceptions then those statements must be written in try block and hence try block is called **exception monitering block**.
 * When the exception occurs  in try block then PVM comes out of try block and excecutes appropriate except block.
 * When the pvm executes appropriate except block, PVM never goes to try block for executing rest of the statements in try block.
 * Every **try block must be immediately followed by except block**(Otherwise we get error).
 * Every try block **must contain atleast one except block** and it is recommeded to write multiple except blocks for generating multiple user-frinedly error messages.
---

2. <font color='#009900'>**except block:**</font>

* It is the block, in which write **Block of statements generating User-Friendly Error Messages**. In otherwords, **except block supresses the technical error messages and generates user-friendly error Messages** and except block is called **Exception Processing Block**.
> Note:- **Handling the Exception = try block + except block**.
* except block will execute when there is an exception occurs in try block.
* Even though we write multiple except blocks, at any point of time only appropriate except block will execute depends on type of exception occurs in try block.
* The place of writing except block is **after try block and before else block**(if we wrirte else block).
---

3. <font color='#009900'>**else block:**</font>

* It is the block, In which we write **block of statements generating Result of the program**
* else block will **execute when there is no exception exception occurs in try block**
* Writing **else block is optional**
* we write else block **after except block and before finally block**(if we write finally block)
--- 

4. <font color='#009900'>**finally block:**</font>

* It is the block, in which we write **block of statements relinquish (release / close / give-up/ clean-up) the resources ( Files/ Database) which are obtained in try block**.
* finally block **will execute compulsorily irrespective of type of exception occurs or not**
* Writing finally block is **optional**.
* finally block **to be written after else block**
